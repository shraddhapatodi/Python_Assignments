Getting the data of GDP per capita growth for USA as a DataFrame:

```python
indicator1 = {"NY.GDP.PCAP.KD.ZG": "GDPPC"}
gdp_percapita = wbdata.get_dataframe(indicator1, country="USA")
gdp_percapita.head(15)
```

Getting the data of USA Stock Market Return as a DataFrame:

```python
indicator2 = {"GFDD.OM.02":"Stock Market Return"}
stock_market_return = wbdata.get_dataframe(indicator2, country="USA")
stock_market_return.head(15)
```

Getting the data for USA Net Migration as a Dataframe:

```python
indicator3 = {"SM.POP.NETM": "Net Migration"}
Net_Mig = wbdata.get_dataframe(indicator3,country = "USA")
Net_Mig.head(15)
```

Issue with Net Migration Data : The above values are not in percentage.Thus,for the purpose of converting Net Migration data into percentage,we need to take the total Population of USA.

```python
indicator4 ={"SP.POP.TOTL":"Population"}
total_pop = wbdata.get_dataframe(indicator4,country = "USA")
total_pop.head(15)
```

Joining the 2 dataframes:

```python
Mig_Pop = Net_Mig.join(total_pop)
Mig_Pop.head(15)
```

```python
For converting the Net Migration values into percentage values:
```

```python
Mig_Perc = Mig_Pop['Net Migration']/Mig_Pop['Population']*100
Mig_Perc.head(15)
```

```python
type(Mig_Perc)
```

The above is a series.Thus , we need to convert it into a datframe:

```python
Net_Migration_Perc = Mig_Perc.to_frame(name = 'Net Migration')
Net_Migration_Perc.head(15) # Displaying only top 15 values
```

ISSUES : 1. Net Migration data is not yearly.It is given for a period of 5 years and rest of the dataframes have annual data.
         2. There are many NaN values in all the dataframes.We need to clean them!

Now that we have all the dataframes and with similar value formats(data in percentage), we need to clean and combine them.

# Combining Data first by Joining the dataframes:

```python
US_Dataframe = gdp_percapita.join(stock_market_return).join(Net_Migration_Perc)
US_Dataframe
```

# CLEANING THE DATA :

Replacing the NaN values by filling values backward:
Reference taken : https://pandas.pydata.org/pandas-docs/stable/missing_data.html
We can make a copy of the dataframe or else work on the same data:

```python
US_Dataframe.fillna(method='bfill', inplace=True)
US_Dataframe.head(15)
```

# Correlation of GDP with Stock Market Return:

# Correlation of GDP with Net Migration:

```python
US_Dataframe['GDPPC'].corr(US_Dataframe['Net Migration'])
```
